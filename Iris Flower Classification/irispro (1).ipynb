# importing the required libraries
 
 

```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split

## GETTING DATA FROM LOCAL DEVICE
column=['sepal length','sepal width','petal length','petal width','class labels']
df=pd.read_csv('C:\\Users\\Dell\\OneDrive\\Desktop\\coincent\\irisdata.csv',names=column)
df.head(150)
```

```python
df.describe()
```

```python
# ploting data to know the relation 

sb.pairplot(df,hue='class labels')
```

```python


data=df.values
Y=data[:,4]
X=data[:,0:4]
print(X)
```

```python

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print((X_train),len(X_test), (Y_train),len(Y_test))
```

```python

from sklearn.svm import SVC
obj=SVC()
obj.fit(X_train,Y_train)
prediction=obj.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction))
```

```python
# using regression model 
from sklearn.linear_model import LogisticRegression
obj1= LogisticRegression()
obj1.fit(X_train,Y_train)
prediction1=obj1.predict(X_test)
```

```python

print(accuracy_score(Y_test,prediction1))
```
